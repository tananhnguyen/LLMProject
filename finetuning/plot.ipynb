```python
import json
import matplotlib.pyplot as plt
import random

# Load the log file
with open("/home/tananh/llm_subject/finetuning/results/checkpoint-65160/trainer_state.json") as f:
    data = json.load(f)

log_history = data["log_history"]

# Randomly select up to 100 entries
num_samples = min(100, len(log_history))
selected_logs = random.sample(log_history, num_samples)

# Sort selected logs by step for meaningful plots
selected_logs = sorted(selected_logs, key=lambda x: x["step"])

# Extract metrics for plotting
steps = [entry["step"] for entry in selected_logs]
loss = [entry["loss"] for entry in selected_logs]
accuracy = [entry["mean_token_accuracy"] for entry in selected_logs]
grad_norm = [entry["grad_norm"] for entry in selected_logs]
learning_rate = [entry["learning_rate"] for entry in selected_logs]

# Plot Loss
plt.figure(figsize=(10, 5))
plt.plot(steps, loss, marker="o")
plt.title("Training Loss (Random 100 Steps)")
plt.xlabel("Step")
plt.ylabel("Loss")
plt.grid(True)
plt.show()

# Plot Mean Token Accuracy
plt.figure(figsize=(10, 5))
plt.plot(steps, accuracy, marker="o", color="green")
plt.title("Mean Token Accuracy (Random 100 Steps)")
plt.xlabel("Step")
plt.ylabel("Mean Token Accuracy")
plt.grid(True)
plt.show()

# Plot Gradient Norm
plt.figure(figsize=(10, 5))
plt.plot(steps, grad_norm, marker="o", color="orange")
plt.title("Gradient Norm (Random 100 Steps)")
plt.xlabel("Step")
plt.ylabel("Grad Norm")
plt.grid(True)
plt.show()

# Plot Learning Rate
plt.figure(figsize=(10, 5))
plt.plot(steps, learning_rate, marker="o", color="red")
plt.title("Learning Rate (Random 100 Steps)")
plt.xlabel("Step")
plt.ylabel("Learning Rate")
plt.grid(True)
plt.show()
```
